# Line Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y=[10,20,15,25]

plt.plot(x,y,'y:',marker='o')  # NY COLOUR JUST WRITTEN FIRST LETTTER and use of colon  and use of marker="O for marking at sharpen side or turn side"
plt.show()
```

```python
import pandas as pd , seaborn as sns

df = pd.DataFrame({"day" : [1,2,3,4] ,"sales" : [10,20,15,25]})

sns.lineplot(x= "day" , y="sales" , data= df,linestyle="--",marker="o") 

plt.show()
```

# Bar Plot

```python
import matplotlib.pyplot as plt

x=["A","B","C"]
y=[10,20,15]
plt.bar(x,y)
plt.show()
```

# add h for horizontal

```python
plt.barh(x,y, color="green")
plt.show()
```

```python
import pandas as pd , seaborn as sns

df= pd.DataFrame({"cat" : ["A" ,"B","C"],"val" :[10,20,15] })
sns.barplot(x="cat", y="val", data=df)
```

# Violin plot


```python
import seaborn as sns
import pandas as pd

data = {
    "Category" : ["A","A","A","B","B","B","C","C","C"],
    "Score" : [10,12,14,20,22,19,30,28,35]
}

df= pd.DataFrame(data)

sns.violinplot(x="Category", y="Score", data=df, color="skyblue")
```

```python
sns.violinplot(y="Category",x="Score",data=df)
```

```python
sns.violinplot(x="Category",y ="Score", data = df , inner ="quartile")
```
